```python
#import library
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YM2APOgzY2w')
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
period=soup.find_all('p',class_='period-name')
period
```

```python
Period=[]
for i in period:
    Period.append(i.text)
Period
```

```python
discription=soup.find_all('p',class_='short-desc')
discription
```

```python
Discription=[]
for i in discription:
    Discription.append(i.text)
Discription
```

```python
temprature_high=soup.find_all('p',class_='temp temp-high')
temprature_high
```

```python
Temp_high=[]
for i in temprature_high:
    Temp_high.append(i.text)
Temp_high
```

```python
temprature_low=soup.find_all('p',class_='temp temp-low')
temprature_low
```

```python
details=soup.find_all('div', class_='col-sm-10 forecast-text')
details
```

```python
details_forcast=[]
for i in details:
    details_forcast.append(i.text)
details_forcast
```
