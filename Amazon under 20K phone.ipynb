```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
Phone_name=soup.find_all('span',class_='a-size-base-plus a-color-base a-text-normal') 
Phone_name
```

```python
Name=[]
for i in Phone_name:
    Name.append(i.text)
model=Name
model[0:16]
```

```python
Price=soup.find_all('span',class_='a-price-whole')
Price
```

```python
Rate=[]
for i in Price:
    Rate.append(i.text)
Rate[0:16]
```

```python
image=soup.find_all('div',class_='a-section aok-relative s-image-square-aspect')
image[0:16]
```

```python
avg_rating=[]
for i in Rating:
    avg_rating.append(i.text)
    
avg_rating
```
